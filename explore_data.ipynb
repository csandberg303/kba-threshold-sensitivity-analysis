# PreProcessing Workflow
I prepared the file with pine hexes to run in marxanconpy using following steps:
1. Plotted raster of US Ecoareas using ArcMap
2. Selected only raster cells featuring Atlantic Coastal Plain Fall-line Sandhills Longleaf Pine Woodland
3. Converted selected raster cells to plygons using Raster to Polygon ArcMap Tool
4. Used Select by Location tool and chose fragments of shapefile with hexes which overlap with Longleaf Pine polygons
5. Exported resulting selection of hexes as a shapefile

```python
# Read pine hexes
pine_hex = gpd.read_file(os.path.join( 
    'pine_hexes_corrected', 
    'pine_hexes.shp'
))

# Plot pine hexes
pine_hex.plot()
```

```python
# Plot fragment of pine hexes
pine_hex_fragment = gpd.read_file(os.path.join(
    'pine_hex_fragment', 
    'pine_hex_fragment.shp'
))

pine_from_raster_fragment = gpd.read_file(os.path.join(
    'pine_from_raster_fragment', 
    'pine_from_raster_fragment.shp'
))

# Plot pine hexes fragment
fig, ax = plt.subplots(figsize=(12, 12))
pine_from_raster_fragment.plot(ax=ax, color='green')
pine_hex_fragment.plot(ax=ax, color='none', edgecolor='black')
ax.set_title('Atlantic Coastal Plain Fall-line Sandhills Longleaf \
Pine Woodland Ecosystem',
             fontsize=16)
```
